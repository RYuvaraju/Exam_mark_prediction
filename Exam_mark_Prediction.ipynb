## Segregate Dataset into Input X & Output Y

```python
x=dataset.iloc[:,:-1].values
print(x.shape)
print(x)
```

```python
y=dataset.iloc[:,-1]
y
```

## Training Dataset using Linear Regression

```python
model=LinearRegression()
model.fit(x,y)
```
